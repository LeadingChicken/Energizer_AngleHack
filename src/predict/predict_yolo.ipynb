```python
class_count = {
 "beer_bottle":0,
 "beer_keg":0,
 "billboard":0,
 "bivina_logo":0,
 "bucket":0,
 "campain-objects":0,
 "consumer":0,
 "display-stand":0,
 "fridge":0,
 "heineken_logo":0,
 "larue_logo":0,
 "parasol":0,
 "pg_marketer":0,
 "saigon_logo":0,
 "signage":0,
 "staff":0,
 "standee":0,
 "strongbow_logo":0,
 "tent-card":0,
 "tiger_logo":0
}
```

```python
def count_classes(predictions):
    data = predictions['predictions']
    for obj in data:
        class_count[obj['class']]+=1

count_classes(predictions)
class_count
```

```python
class Object():
	def __init__(self,x,y,w,h,name):
		self.xul = x
		self.yul = y
		self.xlr = x+w
		self.ylr = y+h
		self.name = name
		self.xCenter = (self.xul+self.xlr)//2
		self.yCenter = (self.yul+self.ylr)//2

objects = []
beer_logo_class = {'bivina_logo','heineken_logo','larue_logo','saigon_logo','strongbow_logo','tiger_logo','standee','tent-card'}
beer_logo = []
people_class = {'consumer','pg_marketer','staff'}
people = []
items_class = {'beer_bottle','beer_keg','billboard','bucket','campain-objects','display-stand','fridge','parasol','signage'}
items = []
for value in predictions['predictions']:
	item = Object(value['x'],value['y'],value['width'],value['height'],value['class'])
	objects.append(item)
	if item.name in beer_logo_class:
		beer_logo.append(item)
	elif item.name in people_class:
		people.append(item)
	else:
		items.append(item)


```

```python
# Check whether (x,y) inside a rectangle (xul,yul,xlr,ylr)
def check_point_inside_rectangle(xul,yul,xlr,ylr,x,y):
	if x>=xul and x<=xlr and y>=yul and y<=ylr:
		return True
	return False
# output logo of the item or none
def find_items_logo(item,logos):
	for logo in logos:
		# if the logo center is inside that item
		if check_point_inside_rectangle(item.xul,item.yul,item.xlr,item.ylr,logo.xCenter,logo.yCenter):
			return logo.name
	
	return None
```

```python
for item in items:
	print(item.name,find_items_logo(item,beer_logo))
```
